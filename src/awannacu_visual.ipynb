```python
import numpy as np
import matplotlib.pyplot as plt

filepath = "../awannacu_serial_visual.raw"
width = 200
height = 200 

data = np.fromfile(filepath, dtype=np.int32, count=height * width)
image = data.reshape((height, width))


plt.imshow(image, cmap='viridis')
plt.colorbar(label='Visible Pixels')
plt.title("Viewshed Visibility")
plt.xlabel("X Pixel")
plt.ylabel("Y Pixel")
plt.show()
plt.imsave(f"viewshed{height}x{width}.png", image, cmap='viridis')
```
